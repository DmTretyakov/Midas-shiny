```python
import requests      
import numpy as np   
import pandas as pd  
import time          
from bs4 import BeautifulSoup
from datetime import datetime

df = pd.read_excel('http://www.gks.ru/free_doc/new_site/prices/potr/I_ipc.xlsx')

df = df.iloc[ [2] +list(np.arange(4,16)) ]

df.columns = df.iloc[0]

df.drop(2,inplace = True)

df.columns = ['mon'] + [str(x) for x in np.arange(1991,2021)]

df= df.drop([str(x) for x in np.arange(1991,2008)], axis = 1)





df = pd.melt(df,value_vars=[str(x) for x in np.arange(2008,2021)])

df['bas'] =  np.zeros(df.shape[0])
df['inf'] =  np.zeros(df.shape[0])

df.bas[0] = 100

df.dropna(axis='index', how='any',inplace = True)

for i in range(df.shape[0] - 1):
    df.bas[i+1] = df.bas[i] * df.value[i+1] / 100

df

for i in range(df.shape[0] - 12):
    df.inf[i + 12] = df.bas[i + 12] / df.bas[i] - 1

df = df.drop(['value','bas'], axis = 1)

df = df.iloc[12:]




df.to_excel('C:/Users/DIMA/shiny/example/inflation.xlsx')






date = datetime.now().strftime("%d.%m.%Y")
page_link = 'http://www.cbr.ru/currency_base/dynamics/?UniDbQuery.Posted=True&UniDbQuery.mode=1&UniDbQuery.date_req1=&UniDbQuery.date_req2=&UniDbQuery.VAL_NM_RQ=R01235&UniDbQuery.From=01.01.2010&UniDbQuery.To=' + date
response = requests.get(page_link)
response

html = response.content

response = requests.get(page_link)
response

soup = BeautifulSoup(html,'html.parser') 

soup.html.head.title.text

obj = soup.find('table', attrs = {'class':'data'}).find_all('td')

l = []
for i in obj:
    l.append(i.text)

date1 = l[1:len(l):3]
usd1 = l[3:len(l):3]

d = {'date': date1, 'usd': usd1}
df_usd = pd.DataFrame(data=d)

df_usd


df_usd.to_excel('C:/Users/DIMA/shiny/example/usd.xlsx')











date = datetime.now().strftime("%d.%m.%Y")
page_link = 'https://www.cbr.ru/hd_base/mkr/mkr_base/?UniDbQuery.Posted=True&UniDbQuery.From=01.01.2010&UniDbQuery.To='+date+'&UniDbQuery.st=SF&UniDbQuery.ob=OB_MIACR_0&UniDbQuery.Currency=1&UniDbQuery.sk=Dd1_'
response = requests.get(page_link)
response

html = response.content

page_link

soup = BeautifulSoup(html,'html.parser') 

soup.html.head.title.text

obj = soup.find('table', attrs = {'class':'data'}).find_all('td')
obj

l = []
for i in obj:
    l.append(i.text)

date1 = l[:len(l):2]
mia1 = l[1:len(l):2]

date1.reverse()
mia1.reverse()

{'col1': [1, 2], 'col2': [3, 4]}

d = {'date': date1, 'mia': mia1}
df_mia = pd.DataFrame(data=d)

df_mia


df_mia.to_excel('C:/Users/DIMA/shiny/example/miacr.xlsx')
```
